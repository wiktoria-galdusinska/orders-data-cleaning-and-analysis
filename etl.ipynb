# Formatting `orders_` tables

In further analysis we don't need the `CustomerID` column, so we are going to remove it from all `orders_` tables.

```python
def get_month_num(month_name: str) -> int:
    """
    Args
    - month_name: string with name of the month

    Returns
    - int: month number
    """
    months = {
        'January': 1, 'February': 2, 'March': 3, 'April': 4,
        'May': 5, 'June': 6, 'July': 7, 'August': 8,
        'September': 9, 'October': 10, 'November': 11, 'December': 12
    }

    return months.get(month_name)


def parse_order_date(df: pd.DataFrame, month_name: str) -> pd.DataFrame:
    """
    Args
    - df: dataframe with `OrderDate` column
    - month_name: string representing month of order

    Returns
    - dataframe with parsed `OrderDate`
    """

    formats = ['%d-%m-%Y', '%m-%d-%Y', '%Y-%m-%d', '%Y-%d-%m', '%d/%m/%Y', '%m/%d/%Y', '%Y/%m/%d', '%Y/%d/%m']
    
    def parse_single_date(date_str):
        for fmt in formats:
            try:
                dt = pd.to_datetime(date_str, format=fmt)
                if dt.month == get_month_num(month_name):
                    return dt.strftime('%d/%m/%Y')
            # can't parse date using given format
            except ValueError:
                continue
        # can't parse date for any expected format
        return None
        
    df['OrderDate'] = df['OrderDate'].apply(parse_single_date)

    return df

def parse_requested_delivery_date(df: pd.DataFrame) -> pd.DataFrame:
    """
    Args
    - df: dataframe with `OrderDate` and `RequestedDeliveryDate` columns

    Returns
    - dataframe with parsed `RequestedDeliveryDate`
    """
    formats = ['%d-%m-%Y', '%m-%d-%Y', '%Y-%m-%d', '%Y-%d-%m', '%d/%m/%Y', '%m/%d/%Y', '%Y/%m/%d', '%Y/%d/%m']
    
    def parse_single_date(date_str, order_date):
        order_date = pd.to_datetime(order_date, format='%d/%m/%Y')

        potential_delivery_dates = []
        for fmt in formats:
            try:
                dt = pd.to_datetime(date_str, format=fmt)
                if dt >= order_date:
                    potential_delivery_dates.append(dt)
            # can't parse date using given format
            except ValueError:
                continue
        
        if potential_delivery_dates:
            return min(potential_delivery_dates).strftime('%d/%m/%Y')
        
        # can't find a reasonable delivery date
        return None
    
    df['RequestedDeliveryDate'] = df.apply(lambda row: parse_single_date(row['RequestedDeliveryDate'], row['OrderDate']), axis=1)
    return df


def parse_dates(df: pd.DataFrame, month_name: str) -> pd.DataFrame:
    """
    Args
    - df: dataframe with `OrderDate` and `RequestedDeliveryDate` columns
    - month_name: string representing month of order

    Returns
    - dataframe with parsed `OrderDate` and `RequestedDeliveryDate` columns
    """

    df = parse_order_date(df, month_name)
    df = parse_requested_delivery_date(df)

    return df

def orders_table_format(df: pd.DataFrame, month_name: str) -> pd.DataFrame:
    """
    Args
    - df: dataframe with `OrderDate` and `RequestedDeliveryDate` columns
    - month_name: string representing month of order

    Returns
    - dataframe with parsed `OrderDate` and `RequestedDeliveryDate` columns, removed `CustomerName`, `CustomerID` and `Comments` columns, standardized encoding in remaining columns
    """

    df = df.copy()

    if month_name == 'July':
        df = df.rename(columns = {'Column1': 'ProductID'})
        df['ProductID'] = df['ProductID'].str.replace('PRODUCT-', '')

    elif month_name == 'August':
        df.columns = ['OrderID', 'CustomerID', 'CustomerName', 'ProductID', 'Quantity', 'OrderDate', 'RequestedDeliveryDate', 'Comments']

    elif month_name == 'November':
        df['OrderID'] = df['OrderID'].str.replace('ORD', 'O')
        df['CustomerID'] = df['CustomerID'].str.replace('CUST', 'C0')
        df['ProductID'] = df['ProductID'].str.replace('PROD', 'P0')

    df = df.drop(columns = ['CustomerName', 'CustomerID', 'Comments'], errors='ignore')
    df = parse_dates(df, month_name)

    return df
```

```python
orders_dict = {
    "January": orders_January,
    "February": orders_February,
    "March": orders_March,
    "April": orders_April,
    "May": orders_May,
    "June": orders_June,
    "July": orders_July,
    "August": orders_August,
    "September": orders_September,
    "October": orders_October,
    "November": orders_November
}

orders = []
for month, orders_df in orders_dict.items():
    orders.append(orders_table_format(orders_df, month))

orders = pd.concat(orders, ignore_index=True)
orders
```

* Notice that `Issue` is always the same. We could skip that column, because it wouldn't matter in further analysis.

```python
complaints['ProductID'] = complaints['ProductID'].str.replace(r'^00', '', regex=True)

complaints.head(15)
```

```python
print(complaints['Issue'].nunique())
print(complaints['Issue'].isna().sum())
```

The code above confirms, that there is only one unique value and zero missing values in the `Issue` column, which means we could skip it.

```python
complaints = complaints.drop(columns='Issue', errors='ignore')
```

```python
(deliveries['DeliveryStatus'] == 'Damaged').sum()
```

All Damaged packages (with `DeliveryStatus == Damaged`) are listed in `complaints` table. We could join `deliveries` and `complaints` to add a column `Carrier` to `complaints` to have easier access to that information later in the analysis.

```python
complaints = complaints.drop(columns='Carrier', errors='ignore')

complaints = complaints.merge(
    deliveries[['OrderID', 'Carrier']],
    on='OrderID',
    how='left'
)
complaints
```

In further analysis, we are going to need information about the products in that table. We could also join `product_master` and `complaints` to add columns `ProductName`, `ProductCategory` and `ProductSubCategory` to `complaints` to have easier access to that information later in the analysis.

```python
complaints = complaints.drop(columns=['ProductName', 'ProductCategory', 'ProductSubCategory'], errors='ignore')

complaints = complaints.merge(
    product_master[['ProductID', 'ProductName', 'ProductCategory', 'ProductSubCategory']],
    on='ProductID',
    how='left'
)
complaints
```
